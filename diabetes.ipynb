# Train & Test & Validation 

```python
data_input =DD.drop('Outcome', axis = 'columns')
data_output = DD['Outcome']
X_train, X_test, y_train, y_test = train_test_split(data_input, data_output, test_size=0.30, random_state=0)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.30, random_state=0)
print('X_train:', X_train.shape)
print('y_train:', y_train.shape)
print('------------------------')
print('X_val:', X_val.shape)
print('y_val:', y_val.shape)
print('------------------------')
print('X_test:', X_test.shape)
print('y_test:', y_test.shape)
```

# Modeling 

# Reusable function

```python
def calculate_accuracy_diabetes(model, X_train, y_train, X_val, y_val):
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    y_pred_val = model.predict(X_val)
    acc_train = accuracy_score(y_train, y_pred_train)
    acc_val = accuracy_score(y_val, y_pred_val)
    return acc_train, acc_val
```

# KNN

## Find best value for K

```python
from sklearn.neighbors import KNeighborsClassifier
Ks=list(range(1,20,2))
train_acc = []
val_acc = []
for k in Ks:
    model = KNeighborsClassifier(n_neighbors = k)
    knn_accuracy = calculate_accuracy_diabetes(model, X_train, y_train, X_val, y_val)
    train_acc.append (knn_accuracy[0])
    val_acc.append(knn_accuracy[1])
KNN_result = pd.DataFrame({'k':Ks,'train_accuracy':train_acc,'validation':val_acc})
KNN_result.set_index('k')
```

## Represent values of knn 

```python
KNN_result.plot(x='k', y=['train_accuracy', 'validation'], figsize=(5, 5))
plt.grid(axis='both')
```

According to the previous plot, we select `n_neighbours = 7` as the best value for `Knn`

# best value of k is 7

```python
model_knn = KNeighborsClassifier(n_neighbors = 7)
knn_acc = calculate_accuracy_diabetes(model_knn, X_train, y_train, X_val, y_val)
print(knn_acc)
```

# Random forest 

## Find best value for estimator

```python
# best estimator at 320
rf_estimators =[80,160,240,320,400,480,560,640,720]
train_acc=[]
val_acc=[]
for estimator in rf_estimators:
    model_RF = RandomForestClassifier(n_estimators=int(estimator), max_depth=2, random_state=0)
    RF_acc=calculate_accuracy_diabetes(model_RF, X_train, y_train, X_val, y_val)
    train_acc.append (RF_acc[0])
    val_acc.append(RF_acc[1])
RF_result = pd.DataFrame({'number_of_estimators':rf_estimators,'train_accuracy':train_acc,'validation':val_acc})
RF_result.set_index('number_of_estimators')  
```

## Represent estimator for random forest 

```python
RF_result.plot(x='number_of_estimators', y=['train_accuracy', 'validation'], figsize=(8, 6))
plt.grid(axis='both')
plt.xticks(rf_estimators)
plt.show()
```

## Find best value for max_depth

```python
rf_max_depth =list(range(1,10))
train_acc=[]
val_acc=[]
for max_depth in rf_max_depth:
    model_RF = RandomForestClassifier(n_estimators=320, max_depth=int(max_depth), random_state=0)
    RF_acc=calculate_accuracy_diabetes(model_RF, X_train, y_train, X_val, y_val)
    train_acc.append (RF_acc[0])
    val_acc.append(RF_acc[1])
RF2_result = pd.DataFrame({'max_depth':rf_max_depth,'train_accuracy':train_acc,'validation':val_acc})
RF2_result.set_index('max_depth')   
```

## Represent max depth  for random forest 

```python
RF2_result.plot(x='max_depth', y=['train_accuracy', 'validation'], figsize=(8, 6))
plt.grid(axis='both')
plt.xticks(rf_max_depth)
plt.show()
```

According to the previous plots, we select `n_esimator = 320` and `max_depth = 4 `as the best value for `Random Forest`

# best values n_estimator at 320 and max_depth at 4

```python
model_RF = RandomForestClassifier(n_estimators=320,max_depth=4,random_state=0)
RF_acc=calculate_accuracy_diabetes(model_RF, X_train, y_train, X_val, y_val)
print(RF_acc)
```

# Decision Tree

## Find best value for max_depth

```python
from sklearn.tree import DecisionTreeClassifier
dt_max_depth =list(range(1,20))
train_acc=[]
val_acc=[]
for max_depth in dt_max_depth:
    model_DT= DecisionTreeClassifier(max_depth=int(max_depth), random_state=0)
    DT_acc=calculate_accuracy_diabetes(model_DT, X_train, y_train, X_val, y_val)
    train_acc.append (DT_acc[0])
    val_acc.append(DT_acc[1])
DT_result = pd.DataFrame({'max_depth':dt_max_depth,'train_accuracy':train_acc,'validation':val_acc})
DT_result.set_index('max_depth') 
```

## Represent max_depth for decision tree 

```python
DT_result.plot(x='max_depth', y=['train_accuracy', 'validation'], figsize=(10, 7))
plt.grid(axis='both')
plt.xticks(dt_max_depth)
plt.show()
```

## Find best value for critertion

```python
criterion_values = ['gini', 'entropy']
train_acc=[]
val_acc=[]
for criterion in criterion_values:
    Dt_model = DecisionTreeClassifier(max_depth=1, criterion=criterion, random_state=0)
    DT_acc=calculate_accuracy_diabetes(Dt_model, X_train, y_train, X_val, y_val)
    train_acc.append (DT_acc[0])
    val_acc.append(DT_acc[1])
DT2_result = pd.DataFrame({'criterion':criterion_values,'train_accuracy':train_acc,'validation':val_acc})
DT2_result.set_index('criterion') 
```

According to the previous plots, we select  `max_depth = 1 `as the best value for `Decision Tree`

# best values  max_depth at 1 and criterion each gini & entropy as It is equal 

```python
## first with gini 
model_DT1 = DecisionTreeClassifier(max_depth=1, criterion='gini', random_state=0)
DT1_acc=calculate_accuracy_diabetes(model_DT1, X_train, y_train, X_val, y_val)

## second with entropy
model_DT2 = DecisionTreeClassifier(max_depth=1, criterion='entropy', random_state=0)
DT2_acc=calculate_accuracy_diabetes(model_DT2, X_train, y_train, X_val, y_val)

## print value
print(DT1_acc)
print('**********************************************')
print(DT2_acc)
```
